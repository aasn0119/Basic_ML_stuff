#### `Functions`

```python
def printPokemons(numbers):
    pokeman_maps = {
        1: 'Bulbasour',
        2: 'Pikachu',
        3: 'Chalamendor'
    }

    # Loop...!
    for i in numbers:
        if pokeman_maps.get(i) is None:
            print("Pokemon Not found..ðŸ˜…")
        else:
            print(pokeman_maps[i])
       
    return
```

```python
numbers = [int(x) for x in numbers.split()]
printPokemons(numbers)
```

#### `3.Data Visualization`

```python
import matplotlib.pyplot as plt 
```

```python
x = np.array([1,2,3,4,5])
y = x**3 + 3
```

```python
plt.style.use('seaborn')
```

```python
print(x,y)
```

```python
plt.plot(x,y + x, label='Stock Price')
plt.plot(x,y**x, label='Nefty Price')
plt.legend()
plt.xlabel('TimeStamp')
plt.ylabel("Price")
plt.title("My Line Plot")
plt.show()
```

```python
plt.scatter(x,y + x, label='Stock Price')
plt.scatter(x,y - x, label='Nefty Price')
plt.legend()
plt.xlabel('TimeStamp')
plt.ylabel("Price")
plt.title("My Scatter Plot")
plt.show()
```
